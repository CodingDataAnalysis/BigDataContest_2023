```python
### xml을 DataFrame으로 변환하기 ###
from os import name
import xml.etree.ElementTree as et
import pandas as pd
import bs4
import requests
from lxml import html
from urllib.parse import urlencode, quote_plus, unquote

## 각 컬럼 값 ## (포털 문서에서 꼭 확인하세요)

#bs4 사용하여 item 태그 분리
url = 'http://apis.data.go.kr/1480523/StobrService/getStobrInfo'
water_dfs = []
k = 0
while True:
    if k==7:
        break
    k+=1


    # 각 행의 컬럼, 이름, 값을 가지는 리스트 만들기
    row_list = [] # 행값
    name_list = [] # 열이름값
    value_list = [] #데이터값


    # xml 안의 데이터 수집

    params ={'serviceKey' : 'Bwp6imnhQmGs7i9sWG6GNxafF5jgljK4kx20H8UHeoxQ/hO9AlT071AAC+JjJjE3QsoTLxGijjCx1JKNu8WSuA==', 'numOfRows' : '10000', 'pageNo' : f'{k}', 'resultType' : 'xml', 'startYear' : '2019', 'endYear' : '2019', 'adivi_cd' : '', 'city_nm' : '', 'citicoungu_nm' : '', 'adivi_doung_nm' : '', 'adivi_ni_nm' : '' }

    response = requests.get(url, params=params)
    content = response.content
    xml_obj = bs4.BeautifulSoup(content,'lxml-xml')
    rows = xml_obj.findAll('item')
    print(rows)
    """
# 컬럼 값 조회용
columns = rows[0].find_all()
print(columns)
"""

    for i in range(0, len(rows)):
        columns = rows[i].find_all()
        #첫째 행 데이터 수집
        for j in range(0,len(columns)):
            if i ==0:
                # 컬럼 이름 값 저장
                name_list.append(columns[j].name)
            # 컬럼의 각 데이터 값 저장
            value_list.append(columns[j].text)
        # 각 행의 value값 전체 저장
        row_list.append(value_list)
        # 데이터 리스트 값 초기화
        value_list=[]

    #xml값 DataFrame으로 만들기
    water_df = pd.DataFrame(row_list, columns=name_list)
    water_dfs.append(water_df)


# print(water_df.head(19))

#DataFrame CSV 파일로 저장
# water_df.to_csv('water.csv', encoding='utf-8')
```
